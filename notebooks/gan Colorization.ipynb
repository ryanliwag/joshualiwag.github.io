```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
sample_image = cv2.imread("data/fire-force1.png", cv2.IMREAD_GRAYSCALE)
```

```python
def show_image(image):
    plt.figure(figsize=(5,5))
    plt.axis("off")
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.show()   
```

```python
show_image(sample_image)
```

```python
#laplace conversion
laplace=cv2.Laplacian(GB_image, cv2.CV_16S, 40)
laplace[laplace < 0] = 0
laplace=cv2.convertScaleAbs(laplace)
show_image(laplace)

#binary (Otsu thresholding)
ret, thr=cv2.threshold(laplace, 0, 255, cv2.THRESH_OTSU)
show_image(thr)
t, l=cv2.connectedComponents(thr)
print(t)
```
